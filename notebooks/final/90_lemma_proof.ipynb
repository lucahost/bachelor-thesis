```python
import networkx as nx

import matplotlib.pyplot as plt

# proof that isomorph graph has same wiener index

G = nx.Graph()
G.add_nodes_from([1,2,3,4,5])
G.add_edges_from([(1,2),(1,3),(1,4),(1,5),(2,3),(2,4),(2,5),(3,4),(3,5),(4,5)])

H = nx.Graph()

H.add_nodes_from([1,2,3,4,5])
H.add_edges_from([(1,2),(1,3),(1,4),(1,5),(2,3),(2,4),(2,5),(3,4),(3,5),(4,5)])

fig, ax = plt.subplots(1, 2)

ax[0].set_title('G_1, Wiener Index = 10.0')
ax[1].set_title('G_2, Wiener Index = 10.0')

nx.draw(G, ax=ax[0], with_labels=True, )
nx.draw_random(H, ax=ax[1], with_labels=True)

print(f'G and H are isomorph: {nx.is_isomorphic(G,H)}')

print(f'Wiener Index for G: {nx.wiener_index(G)}')
print(f'Wiener Index for H: {nx.wiener_index(H)}')
```

```python
# now proof, that 2 graphs with the same wiener index dont have to be isomorphic

G = nx.Graph()

G.add_nodes_from([1, 2, 3, 4, 5])

G.add_edges_from([(1, 2), (1, 3), (1, 4), (1, 5), (2, 3),
                 (2, 4), (2, 5), (3, 4), (3, 5), (4, 5)])

H = nx.Graph()

H.add_nodes_from([1, 2, 3, 4])

H.add_edges_from([(1, 2), (2, 3), (3, 4)])

fig, ax = plt.subplots(1, 2)

ax[0].set_title('G_1, Wiener Index = 10.0')
ax[1].set_title('G_3, Wiener Index = 10.0')

nx.draw(G, ax=ax[0], with_labels=True, )
nx.draw_random(H, ax=ax[1], with_labels=True)

print(f'G and H are isomorph: {nx.is_isomorphic(G,H)}')

print(f'Wiener Index for G: {nx.wiener_index(G)}')
print(f'Wiener Index for H: {nx.wiener_index(H)}')
```
